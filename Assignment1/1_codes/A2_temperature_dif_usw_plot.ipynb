```python
#Note: The dataset should be in the same fodler while executingfrom datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from pylab import rcParams
rcParams['figure.figsize'] = 15, 5
```

```python
df = pd.read_csv("A2_temperature_dif_usw_sol1", header = None, delimiter="\t")
df[0] = df[0].apply(lambda x: x.split('-')[1])
df[0] = df[0].apply(lambda x : x[0:4]+"-"+x[4:6]+"-"+x[6:])
df.columns = ['date', 'temp_dif']
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.plot(df['date'], df['temp_dif'])
plt.title("Temperature difference")
plt.xlabel("Date")
plt.ylabel("Degree celcius")
```
