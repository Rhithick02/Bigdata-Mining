```python
# Note: Dataset should be in the same directory while executingfrom datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from pylab import rcParams
rcParams['figure.figsize'] = 15, 5
```

```python
df = pd.read_csv("A2_temperature_dif_worldwide", header = None, sep="\t")
df.columns = ['date',  'temp_dif']
df
```

```python
df['date'] = df['date'].astype('str')
```

```python
df['date'] = df['date'].apply(lambda x: x[:4]+"-"+x[4:6]+"-"+x[6:])
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.plot(df['date'], df['temp_dif'])
plt.title("Worldwide Temperature avg.")
plt.xlabel("Date")
plt.ylabel("Degree celcius")
```
